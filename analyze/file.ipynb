```python
import networkx as nx

# Define the file paths
input_file = f'{ROOT}/graph.gpickle'
output_file = f'{ROOT}/graph_edge.pickle'

# Load the graph from the .gpickle file
graph = nx.read_gpickle(input_file)

# Extract the edges
edges = list(graph.edges(data=True))

# Save the edges to a new .pickle file
with open(output_file, 'wb') as f:
    pickle.dump(edges, f)

print(f"Edges have been saved to {output_file}")
```
