```python
# find a user node pair
def find_a_user_pair(tm):
    for i in range(0, tm.shape[0]):
        for j in range(0, tm.shape[1]):
            if tm[i, j] > 0:
                return i, j
    return None

src, dst = find_a_user_pair(train_tm_0)
print(src, dst)
```
